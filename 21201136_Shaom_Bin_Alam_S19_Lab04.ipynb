```python
class Patient:
  def __init__(self, id, name, age, bg):
    self.next = None
    self.prev = None
    self.id = id
    self.name = name
    self.age = age
    self.bloodgroup = bg

class WRM:
  def __init__(self):
    self.head=Patient(None, None, None, None)
    self.head.next = self.head
    self.head.prev = self.head
    self.tail = self.head

  def RegisterPatient(self,id, name, age, bloodgroup):
    newNode = Patient(id, name, age, bloodgroup)
    newNode.next = self.head
    newNode.prev = self.tail
    self.tail.next = newNode
    self.tail = self.tail.next
    self.head.prev = self.tail
    print()
    print("Patient Successfully Registered", end = "\n")
    print()


  def ServePatient(self):
    if self.head.next != self.head:
      delete = self.head.next
      prev = self.head
      next = prev.next.next
      prev.next = next
      next.prev = prev

      print()
      print(f"Currently Serving {delete.name}")
      print()
    else:
      print()
      print("No Patient is Currently in The waiting Room")
      print()

  def CancelAll(self):
    if self.head.next != self.head:
      self.head.next = self.head
      self.head.prev = self.head
      print()
      print("All Appointments Have Been Cancelled")
      print()
    else:
      print()
      print("No Patient is Currently in The Waiting Room")
      print()

  def CanDoctorGoHome(self):
    if self.head.next != self.head:
      print()
      print("No")
      print()
    else:
      print()
      print("Yes")
      print()
  def ShowAllPatient(self):
    if self.head.next != self.head:
      tail = self.head.next
      c = 1
      while tail != self.head:
        print(f"{c}. {tail.name}")
        c += 1
        tail = tail.next
    else:
      print()
      print("No Patient is Currently in The Waiting Room")
      print()

obj=WRM()

while True:
  user_input = int(input("""Please select an option:
                      1. Register Patient
                      2. Serve Patient
                      3. Cancel All
                      4. Can Doctor Go Home?
                      5. Show All Patient
                      6. Exit
                      """))
  if user_input == 1:
    id = int(input("Please Enter The ID of The Patient: "))
    name = str(input("Please Enter The Name of The Patient: "))
    age = int(input("Please Enter The Age of The Patient: "))
    bloodgroup = str(input("Please Enter The Blood-Group of The Patient: "))
    obj.RegisterPatient(id, name, age, bloodgroup)
  elif user_input == 2:
    obj.ServePatient()
  elif user_input == 3:
   obj.CancelAll()
  elif user_input == 4:
    obj.CanDoctorGoHome()
  elif user_input == 5:
    obj.ShowAllPatient()
  elif user_input == 6:
    print("Thank You")
    break
  else:
    print("""Invalid Input
        Try Again""")
```
